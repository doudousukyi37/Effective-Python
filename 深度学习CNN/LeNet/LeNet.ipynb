```python
# super类的作用
class Parent:
    def __init__(self):
        print("Parent's __init__ method")

    def say_hello(self):
        print("Hello from Parent")

class Child(Parent):
    def __init__(self):
        super().__init__()  # 调用父类的构造函数
        print("Child's __init__ method")

    def say_hello(self):
        super().say_hello()  # 调用父类的say_hello方法
        print("Hello from Child")

# 创建Child类的实例
child = Child()
child.say_hello()
```

在PyTorch框架中，`transforms.Normalize` 函数用于对图像数据进行标准化处理，即将图像的每个通道减去均值并除以标准差，以使数据的分布更加标准化，从而提高模型训练的效果。

`transforms.Normalize((0.5,0.5,0.5),(0.5,0.5,0.5))` 中的两个元组分别代表了三个通道（通常是RGB）的均值（mean）和标准差（std）。

具体来说：

- 第一个元组 `(0.5,0.5,0.5)` 表示每个通道的均值。在这里，每个通道的均值都是0.5，意味着在标准化之前，图像的每个像素值将被减去0.5。

- 第二个元组 `(0.5,0.5,0.5)` 表示每个通道的标准差。在这里，每个通道的标准差都是0.5，意味着在减去均值之后，结果将被除以0.5。
